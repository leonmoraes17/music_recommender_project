# creating Item-User interaction 
the matrix will contain 30k+ track ids in rows and 96k+ columns which represent user_ids and the value of each cell will be the playcount. 
since this will be a very large matrix, we will first reduce the size of the matrix by converting the track ids and users_id from String to integers to reduce the DF size
we will use the dask.dataframe.categorize() function which will help in this solution . 


```python
import numpy as np
from scipy.sparse import csr_matrix
```

```python
# Step 1  : Load the df
df = dd.read_csv(user_data_path)
# Step 2 : ensure playcount is numberic
df['playcount'] = df['playcount'].astype(np.float64)
df = df.categorize(columns=['user_id', 'track_id'])
#In Dask, df.categorize() is used to convert object or string columns to categorical types. This can significantly improve performance, 
# especially during operations like groupby(), join(), or merge().

# Step 3 : Convert user_id and track_id to numberic( categorical) values , basically each track_id and user_id is given a numberical(categorixal) value which is done by the categorize function on top
user_mapping = df['user_id'].cat.codes
track_mapping = df['track_id'].cat.codes

```

```python
len(user_mapping)
```

```python
df= df.assign(user_idx = user_mapping, 
              track_idx = track_mapping)
```

```python
df
# two new columns are created showing the categorical value of each user and track id
```

```python
n_rows = df['user_idx'].count()
print(n_rows.compute())
```

```python
# step 4 : compute the interaction matrix
interaction_array = df.groupby(['track_idx', 'user_idx'])['playcount'].sum().reset_index()
interaction_array
```

```python
interaction_array = interaction_array.compute()
interaction_array
```

```python
# creating a sparse matrix 
row_indices = interaction_array['track_idx']
col_indices = interaction_array['user_idx']
values = interaction_array['playcount']

#building the matrix
n_tracks = unique_tracks
n_users = unqiue_users

sparse_matrix = csr_matrix((values, (row_indices, col_indices)), shape= (n_tracks, n_users))
print("Sparse matrix shape", sparse_matrix.shape)
```

```python
def collaborative_recommendation(song_name, dask_data, filtered_data, interaction_matrix, k=5 ):
    song_row = filtered_data[filtered_data['name'] == song_name]
    print(song_row)

    #track_id of song name
    input_track_id = song_row['track_id'].values.item()
    print(input_track_id)
    #index value of track id
    ind= np.where(dask_data['track_id'].cat.categories == input_track_id)[0].item()
    print(ind)
    # fetch the input vector
    input_array = interaction_matrix[ind]
    # get similarity scores
    similarity_scores = cosine_similarity(input_array, interaction_matrix)
    similarity_scores = similarity_scores.ravel()
    
    #get top k recommendations
    recommendations_track_ids = dask_data['track_id'].cat.categories[np.argsort(similarity_scores)[-k-1:][::-1]]
    print(recommendations_track_ids)
    top_scores = np.sort(similarity_scores)[-k-1:][::-1] #getting score parameter to sort data based on most similar score while merging df 
    temp_df = pd.DataFrame({"track_id":recommendations_track_ids.tolist(),
                            "score":top_scores})
    print(temp_df)
    top_k_songs = (
        filtered_data.loc[filtered_data['track_id'].isin(recommendations_track_ids)]
        .merge(temp_df, on = "track_id")
        .sort_values( by = 'score', ascending = False)
        .drop(columns = ['track_id', 'score'])
        .reset_index(drop = True)
        

    )
    return top_k_songs


```
